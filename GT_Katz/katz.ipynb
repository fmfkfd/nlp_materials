```python
import nltk
```

```python
corpus = nltk.corpus.gutenberg.sents(u'austen-emma.txt')
```

```python
def get_unigrams(corpus):
    unigrams = {}
    for sentence in corpus:
        for word in sentence:
            word = word.lower()
            if word not in unigrams:
                unigrams[word] = 0
            unigrams[word] += 1
    return unigrams

def sort_dict(d, reverse = True):
    return sorted(d.items(), key = lambda x : x[1], reverse = reverse)

def get_bigrams(corpus):
    bigrams = {}
    for sentence in corpus:
        for index, word in enumerate(sentence):
            if index > 0:                
                word = word.lower()            
                prev = sentence[index - 1].lower()
                if prev not in bigrams:
                    bigrams[prev] = {}
                if word not in bigrams[prev]:
                    bigrams[prev][word]  = 0
                bigrams[prev][word] += 1
    return bigrams
from operator import itemgetter
from ipy_table import *
def get_list_bigrams(bigrams):
    l=[]
    for k,v in bigrams.items():
        for k1,v1 in v.items():
            l.append((k+"_"+k1,v1))
    l=sorted(l,key=itemgetter(1),reverse=True)
    return l
def get_arbitary_k(bi_list):
    k=0
    for l in bi_list:
        k+=l[1]
    #print(k)
    return (k/len(bi_list))
import math
def unigram_probs(unigrams):
    new_unigrams = {}
    N = sum(unigrams.values())
    for word in unigrams:
        new_unigrams[word] =  round(unigrams[word] / float(N), 15)
    return new_unigrams
```

```python
bigrams = get_bigrams(corpus)
bi_list=get_list_bigrams(bigrams)
alpha=0.7
k=2
unigrams = get_unigrams(corpus)
sorted_unigrams = sort_dict(unigrams)
uprobs = unigram_probs(unigrams)
```

```python
def bigram_probs(bigrams):
    new_bigrams = {}
    for word in bigrams:
        if word[0].split('_')[0] in unigrams:
            uni=unigrams[word[0].split('_')[0]]
        new_bigrams[word[0]] =  round(word[1] /uni, 15)
    return new_bigrams
bprobs=bigram_probs(bi_list)
bi_prob=bigram_probs(bi_list[10700:11000])
printed_list=[x[0] for x in bi_prob]
```
